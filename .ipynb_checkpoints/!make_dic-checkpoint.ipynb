```python
import pandas as pd

with open ('dic\pn_en.dic.txt', 'r', encoding='utf-8') as f:
    x = [ii.replace('\n', '').split(':') for ii in f.readlines()]

# PN Tableのデータフレームを作る
posi_nega_df = pd.DataFrame(x, columns = ['基本形', '品詞', 'スコア'])
posi_nega_df = posi_nega_df[~posi_nega_df[['基本形', '品詞']].duplicated()]

# 品詞を日本語にしてみる
for i in range(len(posi_nega_df['品詞'])):
    if posi_nega_df['品詞'][i] == 'v':
        posi_nega_df['品詞'][i] = '動詞'
    elif posi_nega_df['品詞'][i] == 'n':
        posi_nega_df['品詞'][i] = '名詞'
    elif posi_nega_df['品詞'][i] == 'a':
        posi_nega_df['品詞'][i] = '形容詞'
    elif posi_nega_df['品詞'][i] == 'r':
        posi_nega_df['品詞'][i] = '副詞'
posi_nega_df
```

```python
posi_nega_df.to_csv('dic/pn_en.dic.csv')
```
