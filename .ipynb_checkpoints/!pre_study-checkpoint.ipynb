```python
def pn_score(title):
    import pandas as pd
    import numpy as np
    import os
    import glob
    import pathlib
    import re
    import janome
    import jaconv
    import requests
    from bs4 import BeautifulSoup
    import treetaggerwrapper as ttw
    import warnings
    import matplotlib as mpl
    import matplotlib.pyplot as plt
    
    warnings.simplefilter('ignore')
    
    posi_nega_df = pd.read_csv('dic/pn_en.dic.csv', index_col=0)
    
    urlName = "https://www.imsdb.com/scripts/" + title + ".html"
    print(urlName)
    url = requests.get(urlName)
    soup = BeautifulSoup(url.content, "html.parser")
    
    # 先頭のBタグの中身(= タイトル)をscriptに格納
    script = soup.find("b")

    b = soup.findAll("b")
    i = 1

#    title = re.findall('\s*([^\r]*)\r\n', script.text)[0]

#    script = script.next_sibling
#    writer = re.findall('\s*Written\sby\s*([^\r]*)', script.string)[0]

    char_flag = False
    lines = []
    line = ''
    while script.next_sibling != None:
        script = script.next_sibling
        if script.string == b[i].text:
            i += 1
            if re.findall('^\s([^\r])*', script.string) != []:
                char_flag = True
        else:
            s = script.string
            if char_flag:
                j = 0
                while s[0] == ' ':
                    s = s[1:]
                while len(s) > j+3  and  s[j] + s[j+1] + s[j+2] + s[j+3] != '\r\n\r\n':
                    line += s[j].replace('\n', ' ')
                    j += 1
                lines.append(line)
                line = ''
                char_flag = False


    tagger = ttw.TreeTagger(TAGLANG='en')

    word_lists = []
    for i in range(len(lines)):
        for t in tagger.TagText(lines[i]):
            surf = t.split()[0]        # 形態素
            base = t.split()[2]        # 基本形

            # 品詞
            if re.match('V.*', t.split()[1]):
                pos = '動詞'
            elif re.match('WP.*', t.split()[1]) or re.match('PP.*', t.split()[1]):
                pos = '名詞'
            elif re.match('N.*', t.split()[1]):
                pos = '名詞'
            elif re.match('WRB', t.split()[1]) or re.match('EX', t.split()[1]) or re.match('R.*', t.split()[1]):
                pos = '副詞'
            elif re.match('J.*', t.split()[1]) or re.match('PDT', t.split()[1]):
                pos = '形容詞'
            else:
                pos = 'その他'
        
            word_lists.append([i, surf, base, pos])

    word_df = pd.DataFrame(word_lists, columns = ['セリフNo.', '単語', '基本形', '品詞'])
    word_df['品詞'] = word_df['品詞'].apply(lambda x : x.split(',')[0])

    score_result = pd.merge(word_df, posi_nega_df, on = ['基本形', '品詞'], how = 'left')

    result = []
    for i in range(len(score_result['セリフNo.'].unique())):
        temp_df = score_result[score_result['セリフNo.'] == i]
        text = ' '.join(list(temp_df['単語']))
        score = temp_df['スコア'].astype(float).sum()
        score_r = score / temp_df['スコア'].astype(float).count()
        result.append([i, text, score, score_r])

    result_df = pd.DataFrame(result,
                             columns = ['セリフNo.', 'テキスト', '累計スコア', '標準化スコア']
                            ).sort_values(by = 'セリフNo.').reset_index(drop = True)

    result_array = np.nan_to_num(result_df['標準化スコア'].values)

    comp_result = []
    comp = 0
    j = 0
    for i in result_array:
        comp += i
        if j > len(result_array)/25:
            comp_result.append(comp)
            comp = 0
            j = -1
        j += 1
    comp_result.append(comp)

    x = np.arange(len(comp_result))
    plt.plot(x, comp_result)

    pn_change = 0
    for i in range(len(comp_result)-1):
        if comp_result[i]*comp_result[i+1] < 0:
            pn_change += 1

    return pn_change, comp_result
```

```python
from tensorflow import keras
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import RMSprop
import numpy as np
```

```python
x_train = np.array([ 9,  9,  9,  6, 14,  9, 11, 12, 10, 11])
x_train = x_train.reshape(len(x_train), 1)

y_train = np.array([96, 93, 91, 85, 68, 65, 53, 52, 39,  6])
y_train = y_train.reshape(len(y_train), 1)

x_test  = np.array([11,  7,  6,  8])
x_test  = x_test.reshape(len(x_test), 1)

y_test  = np.array([94, 60, 50, 38])
y_test  = y_test.reshape(len(y_test), 1)


for i in range(len(y_train)):
    if y_train[i] < 60:        # rotten(腐敗)
        y_train[i] = 0
    else:        # fresh
        y_train[i] = 1

for i in range(len(y_test)):
    if y_test[i] < 60:        # rotten(腐敗)
        y_test[i] = 0
    else:        # fresh
        y_test[i] = 1


batch_size = 1  # 訓練データをいくつのデータに分けて学習させるか
num_classes = 2 # 分類させる数
epochs = 10 # 訓練データを繰り返し学習させる数

# xを、0-1の実数値にする
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train /= 14
x_test /= 14

# y(目的変数)をKerasで扱いやすいOne-Hot形式に変換する
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)

# モデルの作成
# Sequentialという箱を用意して、そこに深層学習のネットワーク層を追加していく
model = Sequential()
model.add(Dense(512, activation='relu', input_shape=(1,)))
model.add(Dropout(0.2))
model.add(Dense(512, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(num_classes, activation='softmax'))

model.summary()

model.compile(loss='categorical_crossentropy',
              optimizer=RMSprop(),
              metrics=['accuracy'])

# 学習は、scrkit-learnと同様fitで記述できる
history = model.fit(x_train, y_train,
                    batch_size=batch_size,
                    epochs=epochs,
                    verbose=1,
                    validation_data=(x_test, y_test))

# 評価はevaluateで行う
score = model.evaluate(x_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```
